```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_cleaned_dataset.csv')
```

```python
dataset.head()
```

```python
dataset.columns
```

```python
## Drop month, day and year
dataset.drop(['day', 'month', 'year'],axis=1, inplace=True)
```

```python
dataset.head()
```

```python
dataset['Classes'].value_counts()
```

```python
## Encoding
dataset['Classes']=np.where(dataset['Classes'].str.contains("not fire"),0,1)
```

```python
dataset.tail()
```

```python
dataset['Classes'].value_counts()
```

```python
## divide the features into dependent and independent 
X=dataset.drop('FWI',axis=1)
y=dataset['FWI']
```

```python
X.head()
```

```python
y
```

```python
## Train test split
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
## Feature selection based on correlation 
X_train.corr()
```

```python
import seaborn as sns
sns.set()
```

```python
## Check for Multicollinearty
plt.figure(figsize=(12,10))
corr=X_train.corr()
sns.heatmap(corr,annot=True)
plt.show()
```

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
                
    return col_corr        
```

```python
## Thresold--Domain expertise
corr_features=correlation(X_train,0.85)
```

```python
## Drop features when correlation is more than 0.85
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

## Feature Scaling or Standardization

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_scaled=sc.fit_transform(X_train)
X_test_scaled=sc.transform(X_test)
```

```python
X_train_scaled
```

## Box plots to understand Effect of Standard Scaler

```python
plt.subplots(figsize=(15,5))
plt.subplot(1, 2, 1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1, 2, 2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
plt.show()
```

## Linear Regression Model


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
linear = LinearRegression()
linear.fit(X_train_scaled,y_train)
y_pred = linear.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 score", score)
plt.scatter(y_test,y_pred)
plt.show()
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
lasso = Lasso()
lasso.fit(X_train_scaled,y_train)
y_pred = lasso.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 score", score)
plt.scatter(y_test,y_pred)
plt.show()
```

## Cross Validation Lasso

```python
from sklearn.linear_model import LassoCV
lassocv=LassoCV(cv=5)
lassocv.fit(X_train_scaled,y_train)
```

## Ridge Regression Model

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
ridge = Ridge()
ridge.fit(X_train_scaled,y_train)
y_pred = ridge.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 score", score)
plt.scatter(y_test,y_pred)
plt.show()
```

```python
from sklearn.linear_model import RidgeCV
ridgecv=RidgeCV(cv=5)
ridgecv.fit(X_train_scaled,y_train)
y_pred=ridgecv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)

mae = mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 score", score)
plt.scatter(y_test,y_pred)
plt.show()
```
